###Genderwise Age distribution

```python
female_age = df[df['GENDER'] == 'F']['AGE']
male_age = df[df['GENDER'] == 'M']['AGE']

# Create a histogram for each gender
plt.hist([female_age, male_age], bins=17, alpha=0.5, color=['blue', 'green'], label=['Female', 'Male'])

plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.title('Age Distribution by Gender')
plt.show()
```

**Insights**

###Genderwise Trend

```python
def kde(col):
  grid = sns.FacetGrid(df, hue = 'GENDER', height = 4, aspect = 2)
  grid.map(sns.kdeplot, col)
  grid.add_legend()

kde('AGE')
```

**Insights**
- By observing the whole graph, we can see Males have higher values than females in lung cancer
- In between the age of 45 to 65, females are more affected by lung cancer than males

###Checking dependency on Smoking habits

```python
smokers = df[df['SMOKING'] == 1]
non_smokers = df[df['SMOKING'] == 2]

fig, axes = plt.subplots(1, 2, figsize=(9, 5), sharey=True)

# Plot for non-smokers
sns.countplot(ax=axes[0], x='LUNG_CANCER', hue='GENDER', data=non_smokers, palette='Paired')
axes[0].set_title('Lung Cancer in Non-Smokers by Gender')
axes[0].set_xlabel('Lung Cancer')
axes[0].set_ylabel('Count')

# Plot for smokers
sns.countplot(ax=axes[1], x='LUNG_CANCER', hue='GENDER', data=smokers, palette='Paired')
axes[1].set_title('Lung Cancer in Smokers by Gender')
axes[1].set_xlabel('Lung Cancer')

plt.tight_layout()
plt.show()
```

**Insights**
- In the first plot, the data indicates that non-smoker females exhibit higher lung cancer counts than non-smoker males
- In the second plot, the data indicates that smoker males exhibit higher lung cancer counts than smoker-females

###Genderwise distribution of Smoking habits

```python
smokers = df[df['SMOKING'] == 1]
non_smokers = df[df['SMOKING'] == 2]

def kde(col):
  custom_palette = {'F': 'red', 'M': 'green'}
  grid = sns.FacetGrid(df, hue = 'GENDER', height = 4, aspect = 2, palette = custom_palette)
  grid.map(sns.kdeplot, col)
  grid.add_legend()
  plt.title(f'Distribution of {col} by Gender')
  plt.xlabel(col)
  plt.ylabel('Density')
  plt.show()

kde('SMOKING')
```

**Insights**
- 0 to 1.5 indicates smokers
- 1.5 to end indicates non-smokers
- As we can see, Smoking habits are more frequent in males than females

###Checking smoking habits by Gender & Age

```python
def kde_plot_smoking_by_gender():
    # Define custom palette for smokers and non-smokers
    custom_palette = {1: 'red', 2: 'green'}

    # Create subplots for females and males
    fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

    # Plot for females
    sns.kdeplot(ax=axes[0], data=df[df['GENDER'] == 'F'], x='AGE', hue='SMOKING', palette=custom_palette)
    axes[0].set_title('Female Distribution of Smoking')
    axes[0].set_xlabel('AGE')
    axes[0].set_ylabel('Density')
    axes[0].legend(title='SMOKING', labels=['Non-smokers', 'Smokers'])

    # Plot for males
    sns.kdeplot(ax=axes[1], data=df[df['GENDER'] == 'M'], x='AGE', hue='SMOKING', palette=custom_palette)
    axes[1].set_title('Male Distribution of Smoking')
    axes[1].set_xlabel('AGE')
    axes[1].legend(title='SMOKING', labels=['Non-smokers', 'Smokers'])

    # Adjust layout
    plt.suptitle('Distribution of Smoking Habits by Age and Gender', y=1.05)
    plt.tight_layout()
    plt.show()

kde_plot_smoking_by_gender()
```

**Insights**
- Among females aged 30 to 80, non-smokers are more prevalent.
- Among males aged 45 to 90, smoking habits are more common.
